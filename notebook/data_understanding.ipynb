```python
# ─── Import Libraries ───
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set_theme(style='darkgrid')
plt.rcParams['figure.dpi'] = 110

print('Libraries loaded successfully ✅')
```

## Step 1: Load Dataset

We use the **NASA GISS Surface Temperature Analysis (GISTEMP v4)** combined with **NOAA CO₂ (Mauna Loa)** data.

📥 Download links:
- Temperature: https://data.giss.nasa.gov/gistemp/
- CO₂: https://gml.noaa.gov/ccgg/trends/data.html

```python
# Load the raw dataset
# Make sure you have placed the file at: dataset/raw_data/climate_data.csv
df = pd.read_csv('../dataset/raw_data/climate_data.csv')

# Quick look at first 5 rows
print(f'Dataset shape: {df.shape}  →  {df.shape[0]} rows, {df.shape[1]} columns')
df.head()
```

## Step 2: Column Information

```python
# Columns and their data types
df.info()
```

## Step 3: Summary Statistics

```python
# Basic stats: mean, std, min, max, quartiles
df.describe().round(3)
```

## Step 4: Missing Values

```python
# Check null counts per column
null_counts = df.isnull().sum()
null_pct = (null_counts / len(df) * 100).round(2)

missing_report = pd.DataFrame({'Null Count': null_counts, 'Null %': null_pct})
missing_report[missing_report['Null Count'] > 0]
```

```python
# Heatmap of missing values
plt.figure(figsize=(10, 3))
sns.heatmap(df.isnull(), cbar=False, cmap='YlOrRd', yticklabels=False)
plt.title('Missing Value Heatmap', fontsize=13, fontweight='bold')
plt.show()
```

## Step 5: Distribution of Key Variables

```python
# Plot distributions of numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

fig, axes = plt.subplots(1, len(numeric_cols), figsize=(5 * len(numeric_cols), 4))
if len(numeric_cols) == 1:
    axes = [axes]

for ax, col in zip(axes, numeric_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=ax, color='#4361ee')
    ax.set_title(col)

plt.suptitle('Distribution of Numeric Features', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## Step 6: Outlier Detection (Boxplots)

```python
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(5 * len(numeric_cols), 4))
if len(numeric_cols) == 1:
    axes = [axes]

for ax, col in zip(axes, numeric_cols):
    sns.boxplot(y=df[col].dropna(), ax=ax, color='#f4a261')
    ax.set_title(col)

plt.suptitle('Boxplot: Outlier Check', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
